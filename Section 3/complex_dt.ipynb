```python
# 11. Create a tuple of about 5 elements
# Multiply your tuple by 3 and save it to a new 2nd tuple
# Access and print the 12th element from the 2nd tuple
# Sort the 2nd tuple and print it
# Copy 4 specific elements from your 2nd tuple to a new 3nd tuple
# Unpack the 3rd tuple into 4 variables and print these variables
# Create a 4th tuple with single item 50 and print this tuple
# Add the 2nd and 3rd tuple together into a 5th tuple and print the tuple

# Step 1: Create a tuple of about 5 elements
tuple1 = (3, 7, 1, 9, 5)

# Step 2: Multiply your tuple by 3 and save to a new 2nd tuple
tuple2 = tuple1 * 3
print("Tuple2 (repeated):", tuple2)

# Step 3: Access and print the 12th element from the 2nd tuple (index 11, since indexing starts at 0)
print("12th element of tuple2:", tuple2[11])

# Step 4: Sort the 2nd tuple and print it
# Tuples are immutable, so we use `sorted()` to get a sorted list
sorted_tuple2 = sorted(tuple2)
print("Sorted tuple2:", sorted_tuple2)

# Step 5: Copy 4 specific elements from 2nd tuple to a new 3rd tuple
tuple3 = (tuple2[1], tuple2[4], tuple2[7], tuple2[10])
print("Tuple3 (selected items):", tuple3)

# Step 6: Unpack the 3rd tuple into 4 variables and print them
a, b, c, d = tuple3
print("Unpacked values from tuple3:", a, b, c, d)

# Step 7: Create a 4th tuple with single item 50 and print it
tuple4 = (50)
print("Tuple4 (single item):", tuple4)

# Step 8: Add the 2nd and 3rd tuple together into a 5th tuple and print the tuple
tuple5 = tuple2 + tuple3
print("Tuple5 (tuple2 + tuple3):", tuple5)
```

```python
# 12. Create a set of about 3 elements
# Add a list of fruits to the above set and print the result
# Add a car element to your set
# Create a 2nd set with a few odd items
# Save the union of 1st set and 2nd set to a 3rd set
# Pop an element from the 2nd set, and print the remainder of the set
# Clear the 1st set and print the result
# Discard an element, and remove another element from the 3rd set
# Print the remainder of the 3rd set

# Step 1: Create a set of about 3 elements
set1 = {"pen", "book", "notebook"}
print("Set1:", set1)

# Step 2: Add a list of fruits to the above set
fruits = ["apple", "banana", "orange"]
set1.update(fruits)  # .update() adds multiple items from a list or set
print("Set1 after adding fruits:", set1)

# Step 3: Add a car element to your set
set1.add("car")
print("Set1 after adding a car:", set1)

# Step 4: Create a 2nd set with a few odd items
set2 = {"banana", 123, "laptop", "pen"}
print("Set2:", set2)

# Step 5: Save the union of 1st and 2nd set to a 3rd set
set3 = set1.union(set2)
print("Set3 (union of set1 and set2):", set3)

# Step 6: Pop an element from the 2nd set, and print the remainder
popped_item = set2.pop()
print("Popped from set2:", popped_item)
print("Set2 after popping:", set2)

# Step 7: Clear the 1st set and print the result
set1.clear()
print("Set1 after clearing:", set1)

# Step 8: Discard an element, and remove another element from the 3rd set
set3.discard("banana")  # Does nothing if the element isn't there
set3.remove("car")      # Raises error if the element isn't there
print("Set3 after discard and remove:", set3)
```

```python
# 13. Create a dictionary with at least 5 values of different data types
# Print out 1 value
# Replace any one value in your dictionary with your name
# Add your favorite color to the dictionary
# Add a list, tuple or set to your dictionary
# Print a list of the dictionary keys
# Print a list of the dictionary values
# Copy your 1st dictionary into a 2nd dictionary
# Pop an item from the 2nd dictionary, and print the dictionary
# Remove all the elements from the 2nd dictionary and print the result

# Step 1: Create a dictionary with at least 5 values of different data types
my_dict = {
    "age": 25,                      # int
    "height": 5.9,                  # float
    "is_student": True,            # boolean
    "hobbies": ["reading", "gaming"],  # list
    "address": {"city": "Lagos"}   # dictionary
}
print("Original Dictionary:", my_dict)

# Step 2: Print out 1 value
print("Value of 'age':", my_dict["age"])

# Step 3: Replace any one value in your dictionary with your name
my_dict["age"] = "YourName"  # Replace 'age' with a name
print("After replacing age with name:", my_dict)

# Step 4: Add your favorite color to the dictionary
my_dict["favorite_color"] = "Blue"
print("After adding favorite color:", my_dict)

# Step 5: Add a list, tuple, or set to your dictionary
my_dict["favorite_numbers"] = (7, 3, 14)  # adding a tuple
print("After adding a tuple:", my_dict)

# Step 6: Print a list of the dictionary keys
print("Dictionary keys:", list(my_dict.keys()))

# Step 7: Print a list of the dictionary values
print("Dictionary values:", list(my_dict.values()))

# Step 8: Copy your 1st dictionary into a 2nd dictionary
dict2 = my_dict.copy()
print("Copied Dictionary (dict2):", dict2)

# Step 9: Pop an item from the 2nd dictionary, and print the dictionary
popped_item = dict2.pop("is_student")
print("Popped item:", popped_item)
print("dict2 after popping an item:", dict2)

# Step 10: Remove all the elements from the 2nd dictionary and print the result
dict2.clear()
print("dict2 after clearing all items:", dict2)
```
